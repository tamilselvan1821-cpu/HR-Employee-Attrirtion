```python
# MODEL CREATION
```

```python
# creating indepent and dependent vcaraiable
X=data.drop('Attrition',axis=1) # independ value
y=data.Attrition # dependent varaiable
```

```python
X
```

```python
y
```

```python
# preparing trainig and Testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Balance only training data
y_train.value_counts()
```

```python
# Balancing the data
from collections import Counter # importing counter to check of each
from imblearn.over_sampling import SMOTE # for balancing the data
sm=SMOTE()
print(Counter(y_train)) # checking count for each class
X_sm,y_sm=sm.fit_resample(X_train,y_train) # applying sampling on a targert varaiable
print(Counter(y_sm)) # checking the count after sampling for each class
```

```python
# DECSION TREE
```

```python
from sklearn.tree import DecisionTreeClassifier # importing descision tree
dt=DecisionTreeClassifier() # objection creatioin from decision tree
dt.fit(X_sm,y_sm) # traning model
y_hat=dt.predict(X_test) # prediction
y_hat # predicted values
```

```python
# to find overfitting
y_train_predict=dt.predict(X_sm)
y_train_predict
```

```python
##Evaluating the model
from sklearn.metrics import accuracy_score,classification_report,f1_score
```

```python
#training score
acc_train=accuracy_score(y_sm,y_train_predict) #checkkng accuracy
acc_train
#overfit model
```

```python
print(classification_report(y_sm,y_train_predict))
```

```python
# test score
acc_test=accuracy_score(y_test,y_hat) # checking accuracy
acc_test 
```

```python
print(classification_report(y_test,y_hat)) # for testing
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    "criterion":("gini","entropy"),
    "splitter":("best","random"),
    "max_depth":(list(range(1,20))),
    "min_samples_split":[2,3,4],
    "min_samples_leaf":list(range(1,20)),
}
```

```python
tree_clf=DecisionTreeClassifier(random_state=42)
tree_cv=GridSearchCV(tree_clf,params,scoring="f1",n_jobs=-1,verbose=2,cv=3)
tree_cv.fit(X_sm,y_sm)
best_params=tree_cv.best_params_
print(f"Best paramters:{best_params}")
```
